# [436. 寻找右区间](https://leetcode.cn/problems/find-right-interval/){:target="_blank"}
## 描述
### 题干
给你一个区间数组`intervals`其中$intervals[i]=[start_i, end_i]$,且每个$start_i$都**不同**.

区间`i`的**右侧区间**可以记作区间`j`,并且满足$start_j>=end_i$,且$start_j$**最小化**.

返回一个由每个区间`i`**右侧区间** 的最小起始位置组成的数组。如果某个区间`i`不存在对应的**右侧区间**，则下标`i`处的值设为 `-1` 。

### 示例
#### 示例 1：
```
输入：intervals = [[1,2]]
输出：[-1]
解释：集合中只有一个区间，所以输出-1。
```

#### 示例 2:
```
输入：intervals = [[3,4],[2,3],[1,2]]
输出：[-1,0,1]
解释：对于 [3,4] ，没有满足条件的“右侧”区间。
对于 [2,3] ，区间[3,4]具有最小的“右”起点;
对于 [1,2] ，区间[2,3]具有最小的“右”起点。
````

#### 示例 3：
```
输入：intervals = [[1,4],[2,3],[3,4]]
输出：[-1,2,-1]
解释：对于区间 [1,4] 和 [3,4] ，没有满足条件的“右侧”区间。
对于 [2,3] ，区间 [3,4] 有最小的“右”起点。
```

### 提示：
- $1 <= intervals.length <= 2 * 10^4$
- intervals[i].length == 2
- -10^6 <= start_i <= end_i <= 10^6
- 每个间隔的起点都 **不相同**

## 题解


`````python
from typing import List
`````

### 排序加二分查找
可以理解为将所有的`start`取出,然后在start的序列中找出第一个比`end`大的数

`````python

class Solution:
    def findRightInterval(self, intervals: List[List[int]]) -> List[int]:
        start_list = sorted(((s,i) for i,(s,e) in enumerate(intervals)),key=lambda x:x[0])
        ans = []
        for _,e in intervals:
            # 使用二分查找法
            left,right = 0,len(intervals)
            while left < right:
                mid = (left+right) //2
                if e > start_list[mid][0]:
                    left = mid+1
                else:
                    right = mid
            if left == len(intervals):
                ans.append(-1)
            else:
                ans.append(start_list[left][1])
        return ans
            
            
            
        
`````

## 双指针
首先生成两个数组:

- `start_index`: 基于所有区间的起始点从小到大排序。
- `end_index`: 基于所有区间的结束点从小到大排序。并记录当前元素的索引位置
  
我们通过设置两个指针`i` `j`,其初始值为`0`,然后每次循环中取出`end_index[i]`比较`end_index[i]`与`start_index[j]`,如果$end_index[i]>start_index[j]$则将`j+1`,直到找到`j` 使得$$end_index[i] \le start_index[j]$,然后`i+1`,并且在后续的循环中,`j`可以就当前位置继续向后查找而不需要重置回初始位置`0`

`````python
class Solution:
    def findRightInterval(self, intervals: List[List[int]]) -> List[int]:
        start_index = sorted([(x[0],i) for i,x in enumerate(intervals)])
        end_index = sorted([(x[1],i) for i,x in enumerate(intervals)])
        j=0
        ans = [-1]*len(intervals)
        for e,index in end_index:
            while j<len(intervals) and start_index[j][0]<e:
                j+=1
            if j<len(intervals):
                ans[index]=start_index[j][1]
            else:
                break
        return ans
`````
