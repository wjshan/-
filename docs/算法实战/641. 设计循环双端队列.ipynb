# [641. 设计循环双端队列](https://leetcode.cn/problems/design-circular-deque/){:target="_blank"}

## 题干

设计实现双端队列

实现`MyCircularDeque`类:

- `MyCircularDeque(int k)`: 构造函数,双端队列最大为`k`
- `boolean insertFront()`: 将一个元素添加到双端队列的头部.如果操作成功返回`true`,否则返回`false`.
- `boolean insertLast()`: 将一个元素添加到双端队列尾部,如果炒作成功返回`true`,否则返回`false`.
- `boolean deleteFront()`: 从双端队列头部删除一个元素.如果操作成功返回`true`,否则返回`false`.
- `boolean deleteLast()`: 从双端队列尾部删除一个元素.如果操作成功返回`true`,否则返回`false`.
- `int getFront()`: 从双端队列头部获得有个元素,如果双端队列为空,返回`-1`
- `int getRear()`: 获得双端队列的最优一个元素.如果上端队列为空,返回`-1`.
- `boolean isEmpty()`: 若双端队列为空,则返回`true`,否则返回`false`.
- `boolean isFull()`: 若双端队列慢了,则返回`true`,否则返回`false`.


## 示例

### 示例 1

```
输入
["MyCircularDeque", "insertLast", "insertLast", "insertFront", "insertFront", "getRear", "isFull", "deleteLast", "insertFront", "getFront"]
[[3], [1], [2], [3], [4], [], [], [], [4], []]
输出
[null, true, true, true, false, 2, true, true, true, 4]

解释
MyCircularDeque circularDeque = new MycircularDeque(3); // 设置容量大小为3
circularDeque.insertLast(1);			        // 返回 true
circularDeque.insertLast(2);			        // 返回 true
circularDeque.insertFront(3);			        // 返回 true
circularDeque.insertFront(4);			        // 已经满了，返回 false
circularDeque.getRear();  				// 返回 2
circularDeque.isFull();				        // 返回 true
circularDeque.deleteLast();			        // 返回 true
circularDeque.insertFront(4);			        // 返回 true
circularDeque.getFront();				// 返回 4
```

## 提示

- $1 \le k \le 1000$
- $0 \le value \le 1000$
- `insertFront`, `insertLast`,`deleteFront`,`deleteLast`,`getFront`,`getRear`,`isEmpty`,`isFull`调用次数不大于`2000`次

## 题解

### 双向链表模拟双端队列操作

**时间复杂度**: $O(1)$

**空间复杂度**: $O(n)$,每个节点需要被初始化为一个列表.因此虽然是$O(n)$的空间复杂度,但是实际消耗的内存还是比较大的.

![](/assets/images/641.%20设计循环双端队列/2022-08-15-16-02-32.png)

````python
class MyCircularDeque:

    def __init__(self, k: int):
        self.k = k
        self.head = None
        self.tail = None
        self.len = 0
        


    def insertFront(self, value: int) -> bool:
        """
        从头部插入

        Args:
            value (int): 插入值

        Returns:
            bool: true-成功,false-失败(队列已满)
        """
        if self.isFull():
            return False
        if self.head is None and self.tail is None:
            node = [value,None,None]
            self.head = self.tail = node
        else:
            node = [value,None,self.head]
            self.head[1] = node
            self.head = node
        self.len+=1
        return True

    def insertLast(self, value: int) -> bool:
        """
        从尾部插入

        Args:
            value (int): 插入值

        Returns:
            bool: true-成功,false-失败(队列已满)
        """
        if self.isFull():
            return False
        if self.head is None and self.tail is None:
            node = [value,None,None]
            self.head = self.tail = node
        else:
            node = [value,self.tail,None]
            self.tail[2] = node
            self.tail = node
        self.len += 1 
        return True


    def deleteFront(self) -> bool:
        """
        从头部删除一个元素

        Returns:
            bool: true-删除成功,false-删除失败(列表为空)
        """
        if self.isEmpty():
            return False
        self.len -=1
        self.head = self.head[2]
        if self.head is None:
            self.tail = None
        else:
            self.head[1] = None
        return True

    def deleteLast(self) -> bool:
        """
        从尾部删除一个元素

        Returns:
            bool: true-删除成功,false-删除失败(列表为空)
        """
        if self.isEmpty():
            return False
        self.len -=1
        self.tail = self.tail[1]
        if self.tail is None:
            self.head = None
        else:
            self.tail[2] = None
        return True

    def getFront(self) -> int:
        if self.isEmpty():
            return -1
        return self.head[0]


    def getRear(self) -> int:
        if self.isEmpty():
            return -1
        return self.tail[0]


    def isEmpty(self) -> bool:
        return self.len == 0


    def isFull(self) -> bool:
        return self.len >= self.k
````

### 单列表模拟双端队列

从头部插入 指针自减1然后插入值
从尾部插入 在指针指向位置插入值,然后指针自增一

由此可知 `head`指针指向当前头部值的位置,`tail`指针指向当前尾部值的后一位


**时间复杂度**: 各个操作的时间复杂度都是$O(1)$

**空间复杂度**: $O(n)$

![](/assets/images/641.%20设计循环双端队列/2022-08-17-11-43-49.png)

````python
class MyCircularDeque:

    def __init__(self, k: int):
        self.li = [0] * (k+1)
        self.head = self.tail = 0

    def insertFront(self, value: int) -> bool:
        """
        从头部插入

        Args:
            value (int): 插入值

        Returns:
            bool: true-成功,false-失败(队列已满)
        """
        if self.isFull():
            return False
        self.head = (self.head-1)%len(self.li)
        self.li[self.head] = value
        return True

    def insertLast(self, value: int) -> bool:
        """
        从尾部插入

        Args:
            value (int): 插入值

        Returns:
            bool: true-成功,false-失败(队列已满)
        """
        if self.isFull():
            return False
        self.li[self.tail] = value
        self.tail = (self.tail+1) % len(self.li)
        return True


    def deleteFront(self) -> bool:
        """
        从头部删除一个元素

        Returns:
            bool: true-删除成功,false-删除失败(列表为空)
        """
        if self.isEmpty():
            return False
        self.head = (self.head+1)%len(self.li)
        return True

    def deleteLast(self) -> bool:
        """
        从尾部删除一个元素

        Returns:
            bool: true-删除成功,false-删除失败(列表为空)
        """
        if self.isEmpty():
            return False
        self.tail = (self.tail-1)%len(self.li)
        return True

    def getFront(self) -> int:
        """
        从头部读取

        Returns:
            int: 
             - `-1` 列表为空
        """
        if self.isEmpty():
            return -1
        return self.li[self.head]


    def getRear(self) -> int:
        """
        从尾部读取

        Returns:
            int: 
             - `-1` 列表为空
        """
        if self.isEmpty():
            return -1
        return self.li[(self.tail-1)%len(self.li)]


    def isEmpty(self) -> bool:
        return self.head == self.tail


    def isFull(self) -> bool:
        return (self.head - 1) % (len(self.li)) == self.tail
````
